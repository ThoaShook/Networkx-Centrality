##### Using NetworkX library to create and visualize "Flatiron Cohort Mates" graph network

```python
# Create the initial graph
import networkx as nx
G = nx.Graph()
```

```python
# Add nodes
G.add_node('Maia')
```

```python
people = ['Gary', 'Sydney','Amber','Patrick','Christiaan']
for person in people:
    G.add_node(person)
```

```python
# Add edges
G.add_edge('Maia','Sydney')
```

```python
# Create nested data structure and then feed these pairs into the .add_dege() method
relations = {'Gary':['Patrick','Sydney'],'Sydney':['Gary','Patrick','Christiaan','Amber','Maia'],
                'Patrick':['Gary','Christiaan','Sydney','Amber']}
for p1 in relations.keys():
    p2s = relations[p1]
    for p2 in p2s:
        G.add_edge(p1,p2)
```

```python
%matplotlib inline
nx.draw(G, with_labels=True, node_color='#ADFF2F', node_size=2000, alpha=1)
```

##### Create an initial network to visualize a bimodal network of business and yelp reviewers!

```python
# import data stored in the file 'yelp_tip.csv'
import pandas as pd
df = pd.read_csv('yelp_tip.csv')
df_select = df.head(20)
df_select
```

```python
df_select1 = df.tail(20)
df_select1
```

```python
# Create an initial graph
import networkx as nx
G = nx.Graph()
```

```python
# Add nodes
names = {}
node_color = []
for n, person in enumerate(df_select.user_id.unique()):
    name = 'User{}'.format(n)
    names[person] = name
    G.add_node(name)
    node_color.append('cyan')
for n, biz in enumerate(df_select.business_id.unique()):
    name = 'Business{}'.format(n)
    names[biz] = name
    G.add_node(name)
    node_color.append('pink')
```

```python
# iterate through the dataset and create an edge between users and the business they have review:
for row in df_select.index:
    user = df_select['user_id'][row]
    u_name = names[user]
    biz = df_select['business_id'][row]
    b_name = names[biz]
    G.add_edge(u_name, b_name)
```

```python
%matplotlib inline
nx.draw(G, with_labels=True, alpha=1, font_size=9
        , node_size=1000, node_color=node_color)
```

```python
# degree Centrality
nx.degree(G)
```

```python
# the closeness centrality
print(nx.closeness_centrality(G),1)
```

```python
# The Betweeness Metric for a central node
print(nx.betweenness_centrality(G)['User3'])
# The Betweeness Metric for an ostracized node
print(nx.betweenness_centrality(G)['User2'])
```

```python
print(nx.eigenvector_centrality(G)['User1'])
print(nx.eigenvector_centrality(G)['User2'])
```
